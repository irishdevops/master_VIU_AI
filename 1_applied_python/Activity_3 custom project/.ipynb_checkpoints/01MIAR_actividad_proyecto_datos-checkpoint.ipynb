**Importamos paquetes**

```python
# Importamos paquetes a usar

import requests
import json
import datetime
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

**Llamamos la API solo con la información que necesitamos**

```python
# Después de tratar e investigar la base de datos, conseguimos sacar la vista que queremos

pokemon_data_list = []

for i in range(252, 387):                            #El range se corresponde al rango de pokemons de generación II. Que queremos exportar para nuestro experimento.
    url = f"https://pokeapi.co/api/v2/pokemon/{i}/"  #Llamamos la información de cada pokemon iterando sobre la URL.
    response = requests.get(url)
    
    if response.status_code == 200:                   # 200 == API request válida. 400 suele ser la de fallo.
        pokemon_data = response.json()
        pokemon_data_list.append(pokemon_data)
    else:
        print(f"Failed to retrieve data for Pokemon with ID {i}")
```

**Recopilación de Datos - Limpiamos las listas para obtener solo los datos que necesitamos**

```python
Pokemon_Stats_List = [] 


stat_values = []
stat_names = []
pkmns_name = []
pkmtype1 = []
pkmtype2 = []


###Iteramos sobre el archivo JSON recogiendo las columnas que queremos sacar datos
for item in pokemon_data_list:

    stats= item["stats"]
    types= item ["types"]
    pkmn_name = item["name"]


    ###Iteramos sobre la lista  de tipos, concatenando todos los valores en caso de que un pokémon tenga más de dos tipos
   
    pkmtypedict = {}

    for counter,item2 in enumerate(types, start=1):  ## Nota: Técnica aprendida en clase (Incorporando enumerate en vez de utilizar una variable contador)
        pkmtypes = item2["type"]["name"]
        pkmtypedict[f"pkmtype_str{counter}"] = f"{pkmtypes}" ## Nota: Técnica aprendida en clase (usar f"" para concatenar mejor"). Aunque al final modificamos el código y no lo usamos



    ###Iteramos sobre la lista  de stats, obteniendo cada stat por separado como valor único, junto a su nombre.
    for item3 in stats:
            stat_name = item3["stat"]["name"]
            stat_value = item3["base_stat"]
            stat_names.append(stat_name)
            stat_values.append(stat_value)
            pkmns_name.append(pkmn_name)
            pkmtype1.append(pkmtypedict["pkmtype_str1"])
            try: 
               pkmtype2.append(pkmtypedict["pkmtype_str2"])
            except:
               pkmtype2.append(np.nan) ## No podemos hacer un *pass* por que necesitamos que el array sea de la misma longitud
        
    
    

    

data = { 'Pokemon': pkmns_name, 'Type1':pkmtype1, 'Type2':pkmtype2, 'Stat Name': stat_names, 'Stat Value': stat_values}

df = pd.DataFrame(data)
df.head(10)
```

### **Objetivo 4 - Encontrar los pokémon más fuertes teniendo en cuenta las ecuaciones del juego**


*Tenemos en cuenta la fórmula de combate del juego -> Daño = (((2 * Nivel del atacante / 5 + 2) * Potencia * (Estadística de ataque del atacante / Estadística de defensa del defensor)) / 50 + 2) * Modificador de tipo * Modificador adicional*


**La parte que nos interesa a nosotros es la siguiente:  Potencia * (Estadística de ataque del atacante / Estadística de defensa del defensor)) / 50 + 2)**

La potencia es la potencia del ataque que usa el pokemon, pero a nosotros nos interesa ver la estadística de ataque (Ataque o ataque_especial en nuestros datos) y las estadísticas de defensa (defensa y defensa especial). Además, entran en juego la velocidad (El pokemon con mayor velocidad ataca primer) y el HP (Hit Point que puede recibir el pokemon, también conocido como "vida" en la mayoría de juegos)

**A partir de esta parte de la fórmula, vamos a dividir nuestras estrategia de combate en tres tipos.**

**Combinación óptima 1 - Pokemon que ataca primero (mucha velocidad) y tienen mucho ataque. Ecuación utilizada -> Attacker = speed*1.1 + attack**

   Nota: Le damos peso a la velocidad encima del ataque, ya que una buena velocidad es necesaria para atacar primero.
   
**Combinación óptima 2 - Pokemon que ataca primero (mucha velocidad) y tienen mucho ataque especial. Ecuación utilizada -> Special_Attacker = speed*1.1 + special-attack**

   Nota: Le damos peso a la velocidad encima del ataque, ya que una buena velocidad es necesaria para atacar primero.

**Combinación óptima 3 - Pokemon que tiene mucha vida (HP), mucha defensa y mucha defensa especial. Ecuación utilizada -> Tank = HP*1.5 + defense + 0.8* special-defense**

   Nota: Le damos peso a la vida por encima de la defensa y la defensa especial. Esto es por que la vida es general en todos los sentidos ya que no depende si al pokemon le atacan con ataque o ataque especial.


Estas combinaciones permite obetener el pokemon que más se adapta al combate del juego sin saber cuál es tu adversario. Un pokemon por ejemplo, con mucho ataque, pero poca velocidad y una vida intermedia tendrá muchas más probabilidades de ser abatido en el primer turno y ni siquiera llegar a usar ese gran ataque que tiene, por eso, generamos estas ecuaciones que permitirán evaluar el mejor combo de stats dentro de un pokemon. Para hacer esto, añadimos nuevas columnas con nuestra ecuación.



```python
Stats_Battle = df.pivot(index="Pokemon", columns="Stat Name", values="Stat Value").reset_index()
Stats_Battle["Attacker"] = Stats_Battle["speed"] * 1.1 + Stats_Battle["attack"]
Stats_Battle["Special_Attacker"] = Stats_Battle["speed"] * 1.1 + Stats_Battle["special-attack"]
Stats_Battle["Tank"] = Stats_Battle["hp"] * 1.5 + Stats_Battle["defense"] + 0.8 * Stats_Battle["special-defense"]
print(Stats_Battle)
```

**Ahora podemos aplicar visualizaciones y saber cuál es el mejor combo. Cogemos el top 5 de Pokemons y lo dibujamos en una gráfica**

```python

legendary_pokemon = [
    "regirock", "regice", "registeel", "latias", "latios", 
    "kyogre", "groudon", "rayquaza", "jirachi", "deoxys-normal"
] ### Retiramos Pokemons legendarios, ya que nos salen como los valores más altos.

Stats_Battle = Stats_Battle[~Stats_Battle['Pokemon'].isin(legendary_pokemon)]## TRUCO: Como sacar y filtrar una query rápidamente desde una lista. Nota: ~ es el operador de negación. si lo quitamos se quedaría solo con los lefendarios en la lista

top_attackers = Stats_Battle.nlargest(5, 'Attacker') ## TRUCO: Como coger el TOP N de una lista. Para coger el BOTTOM es Stats_Battle.nsmallest(5, 'Tank')
top_special_attackers = Stats_Battle.nlargest(5, 'Special_Attacker')
top_tanks = Stats_Battle.nlargest(5, 'Tank')

# Gráfico de radar. NOTA: Función recogida de la web y retocada posteriormente

def plot_radar_chart(data, title):
    attributes = data.columns[1:7]  # Cogemos las columnas para el radar.
    values = data.values[:, 1:7].tolist()  # Igual con los valores.



    # En Matplot, hace falta poner los ángulos del radar que dibujamos. También incluimos las etiquetas y el título según la función.
    angles = np.linspace(0, 2 * np.pi, len(attributes), endpoint=False).tolist()

   
    fig, ax = plt.subplots(figsize=(2, 2), subplot_kw=dict(polar=True))
    ax.fill(angles, values[0], color='b', alpha=0.25)
    ax.set_xticks(angles)
    ax.set_xticklabels(attributes)
    ax.set_yticklabels([])
    ax.set_title(title)

    
    plt.show() #Importante para enseñar el gráfico


counter=0 ##plot_radar_chart(top_attackers, "Top 5 Attackers")
for i in range(len(top_attackers)):
    counter+=1
    pokemon_data = top_attackers.iloc[i:i+1]
    pokemon_name = pokemon_data['Pokemon'].values[0]
    plot_radar_chart(pokemon_data, f"Stats Radar Chart - Top attacker {counter} - {pokemon_name}")


counter=0 ##plot_radar_chart(top_special_attackers, "Top 5 Special Attackers")
for i in range(len(top_special_attackers)):
    counter+=1
    pokemon_data = top_special_attackers.iloc[i:i+1]
    pokemon_name = pokemon_data['Pokemon'].values[0]
    plot_radar_chart(pokemon_data, f"Stats Radar Chart - Top special attacker {counter} - {pokemon_name}")


counter=0 ##plot_radar_chart(top_tanks, "Top 5 tanks")
for i in range(len(top_tanks)):
    counter+=1
    pokemon_data = top_tanks.iloc[i:i+1]
    pokemon_name = pokemon_data['Pokemon'].values[0]
    plot_radar_chart(pokemon_data, f"Stats Radar Chart - Top tanks {counter} - {pokemon_name}")
```

### **Objetivo 5 - Encontrar correlaciones entre el tipo de pokemon y los stats**

Para realizar este análisis nos damos cuenta de que necesitamos hacer las medias de cada tipo para cada Stat.

```python

df_Types1 = df[['Pokemon','Type1','Stat Name','Stat Value']].copy() 
df_Types2 = df[['Pokemon','Type2','Stat Name','Stat Value']].copy() 
df_Types2['Type2'] = df_Types2['Type2'].replace('', pd.NA)
df_Types2 = df_Types2.dropna(subset=['Type2'])
df_Types1.columns = ['Pokemon', 'Type', 'Stat Name', 'Stat Value']
df_Types2.columns = ['Pokemon', 'Type', 'Stat Name', 'Stat Value']
df_Types= pd.concat([df_Types1, df_Types2], ignore_index=True)
print(df_Types)

average_stats = df_Types.groupby(['Stat Name', 'Type'])['Stat Value'].mean().unstack() #Nos referenciamos a la query del apartado 2 (df_types). NOTA. .unstack() actúa sobre 'Type' para indexar las columnas en base al Type
average_stats = average_stats.reset_index()

#El resultado es una matriz con dimensión Stat Name en las filas. 'Type' de columnas

print(average_stats)

type_colors = {
    'dragon': 'slateblue',
    'steel': 'gray',
    'rock': 'darkred',
    'flying': 'steelblue',
    'psychic': 'deeppink',
    'ice': 'lightblue',
    'fire': 'red',
    'ground': 'saddlebrown',
    'poison': 'purple',
    'dark': 'black',
    'water': 'blue',
    'fighting': 'brown',
    'electric': 'gold',
    'grass': 'green',
    'normal': 'lightgray',
    'ghost': 'indigo',
    'bug': 'limegreen',
    'fairy': 'pink'
} ### Definimos otra vez los colores (Correspondientes a los del juego)

for index, row in average_stats.iterrows(): ### TRUCO: si queremos iterar cada fila obteniendo los valores de cada una por separado, podemos usar iterrows()
    sorted_values = row[1:].sort_values(ascending=False)
    plt.figure(figsize=(8, 6))
    plt.bar(sorted_values.index, sorted_values, color=[type_colors[t] for t in sorted_values.index])
    plt.title(f'Average Value for {row["Stat Name"]}')
    plt.xlabel('Type')
    plt.ylabel('Average Value')
    plt.xticks(rotation=45)
    plt.show()
```

Tras obtener todos los tipos ordenador por tipos de stats podemos sacar algunas conclusiones, como que los pokemon tipo electrico y volador son los más rápidos. O que los pokemons con más defensa son los de tipo acero. Y con más vida los de tipo hielo. Además observamos (Como era de esperar) que los tipo dragón suelen estar en el podio en todos los stats.

### **Objetivo 6 - Encontrar correlaciones entre los stats**

Para realizar este análisis nos damos cuenta de que necesitamos da

```python
df_Types1 = df[['Pokemon','Type1','Stat Name','Stat Value']].copy() 
df_Types2 = df[['Pokemon','Type2','Stat Name','Stat Value']].copy() 
df_Types2['Type2'] = df_Types2['Type2'].replace('', pd.NA)
df_Types2 = df_Types2.dropna(subset=['Type2'])
df_Types1.columns = ['Pokemon', 'Type', 'Stat Name', 'Stat Value']
df_Types2.columns = ['Pokemon', 'Type', 'Stat Name', 'Stat Value']
df_Types = pd.concat([df_Types1, df_Types2], ignore_index=True) #Realizamos el mismo proceso que en el apartado 3, para dividir cada tipo por separado


correlation_df = df_Types.groupby(['Type', 'Stat Name'])['Stat Value'].mean().unstack()


correlation_matrix = correlation_df.corr() # Calculamos la matriz de correlaciones


plt.figure(figsize=(10, 8))# Creamos un mapa de calor
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix of Stat Types')
plt.show()

print("\n Podemos sacar varias conclusiones dentro de este gráfico de correlaciones: \n1. Los pokemons con buen ataque suelen estar equilibrados y tener una buena defensa. En cambio, los pokemon con buen ataque especial, no suelen tener buena defensa \n2. Podemos observar como el paquete entero de defensa, HP y defensa especial tienen correlación positiva entre sí. \n3. Los  Pokemons con defensa alta, suelen ser lentos. Los Pokemons con Ataque especial alto suelen tener buena velocidad.")

 #Hay que tener en cuenta que para el análisis, repetimos los stats de tipos repetidos. Por los que estos van a pronunciar más la matriz de correlaciones resultante
```
